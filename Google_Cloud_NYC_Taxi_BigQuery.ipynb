```python
# Querry dataset to find all taxi rides originating from Carnigie Hall Area

query_carnigie_taxi_rides = """

SELECT pickup_datetime
       , pickup_longitude
       , pickup_latitude
       , dropoff_longitude
       , dropoff_latitude
       , trip_distance
       , total_amount
FROM [nyc-tlc:yellow.trips]
WHERE pickup_longitude BETWEEN -73.980 AND -73.979
  AND pickup_latitude BETWEEN 40.765 AND 40.766;

"""
```

```python
df_carnigie_taxi = pd.read_gbq(query_carnigie_taxi_rides, project_id, private_key=key)
```

```python
df_carnigie_taxi.sort_values('pickup_datetime', inplace=True)
```
